```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Prepare data for plotting
data = {
    'System': ['User Profile', 'User Profile', 'Similarity', 'Similarity', 'Similarity', 'Clustering', 'Clustering', 'KNN', 'KNN', 'NMF', 'NMF', 'Neural Network', 'Neural Network'],
    'Type': ['User 1', 'All Users', 'Toy Story', 'All Users', '610 Movies', 'Cluster 9', 'All Clusters', 'User 1', 'All Users', 'User 1', 'All Users', 'User 1', 'All Users'],
    'Precision': [0.0, 0.0242, 0.0623, 0.0111, 0.0111, 0.0111, 0.0172, 0.0172, 0.0535, 0.0086, 0.1656, 1.0, 0.0791],
    'Recall': [0.0, 0.0023, 0.0052, 0.0007, 0.0007, 0.0012, 0.0011, 0.0331, 0.0045, 0.0165, 0.0257, 1.0, 0.0791],
    'F1-Score': [0.0, 0.0037, 0.0086, 0.0011, 0.0011, 0.0022, 0.0020, 2.364, 0.0076, 4.170, 0.0402, 0.736, 0.0791],
    'RMSE': [2.454, 0.607, 2.437, 2.419, 2.419, 1.620, 1.643, 0.4, 0.484, 0.2, 0.967, 1.0, 0.274]
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)
```

```python
# Set the plotting theme
sns.set_theme(style="whitegrid")
```

```python
# Plot Precision
plt.figure(figsize=(6, 4))
sns.barplot(y='System', x='Precision', hue='Type', data=df, palette='viridis')
plt.title('Comparison of Precision Across Recommender Systems')
plt.xlabel('Precision')
plt.ylabel('Recommender System')
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plot Recall
plt.figure(figsize=(6, 4))
sns.barplot(y='System', x='Recall', hue='Type', data=df, palette='viridis')
plt.title('Comparison of Recall Across Recommender Systems')
plt.xlabel('Recall')
plt.ylabel('Recommender System')
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plot F1-Score
plt.figure(figsize=(6, 4))
sns.barplot(y='System', x='F1-Score', hue='Type', data=df, palette='viridis')
plt.title('Comparison of F1-Score Across Recommender Systems')
plt.xlabel('F1-Score')
plt.ylabel('Recommender System')
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plot RMSE
plt.figure(figsize=(6, 4))
sns.barplot(y='System', x='RMSE', hue='Type', data=df, palette='viridis')
plt.title('Comparison of RMSE Across Recommender Systems')
plt.xlabel('RMSE')
plt.ylabel('Recommender System')
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
