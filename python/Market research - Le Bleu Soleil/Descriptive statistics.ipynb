```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import textwrap

# Read data file
df = pd.read_excel('~/Downloads/data.xlsx')

# Print all column labels
print("Column labels:")
for column in df.columns:
    print(column)
```

```python
# Define the mapping for each column
gender_mapping = {1: 'Male', 2: 'Female'}
age_mapping = {1: '3-5 years old', 2: '6-9 years old', 3: '10-12 years old', 4: '13-15 years old', 5: '16-18 years old'}
frequency_mapping = {1: '3 times per week, 90 minutes per time', 2: '30 minutes per day', 3: '1 hour(s) per day', 4: '2-3 hour(s) per day'}
time_mapping = {1: 'Morning on Saturday, Sunday (8h-11h)', 2: 'After classes during weekdays (Mon-Fri)', 3: 'Evening on Saturday, Sunday (17h-20h)', 4: 'Afternoon on Saturday, Sunday (14h-17h)'}
intensity_mapping = {1: 'Not wet yet', 2: 'Half of the shirt is wet', 3: 'Shirt is wet in front of the chest and behind the back', 4: 'The whole shirt is wet'}
safety_mapping = {1: 'Safe', 2: 'Not injury-prone', 3: 'Minor scrape', 4: 'High impact collision'}
budget_mapping = {1: '100k / session, 90 minutes / session', 2: '500k-600k / session, 8 sessions in total, 90 minutes / session', 3: '1.2m-1.4m / 3 months, 24 sessions in total, 90 minutes / session', 4: '2.2m-2.6m / 6 months, 48 sessions in total, 90 minutes per session'}

# Apply the mapping to each column
df['Gender'] = df['Gender'].map(gender_mapping)
df['Age'] = df['Age'].map(age_mapping)
df['Frequency'] = df['Frequency'].map(frequency_mapping)
df['Time'] = df['Time'].map(time_mapping)
df['Intensity'] = df['Intensity'].map(intensity_mapping)
df['Safety'] = df['Safety'].map(safety_mapping)
df['Budget'] = df['Budget'].map(budget_mapping)
```

```python
# Calculate descriptive statistics for each column
descriptive_stats = df.describe()

# Print the descriptive statistics
print(f'Descriptive statistics')
print()
for column in df.columns:
    print(f'{column}:')
    print(descriptive_stats[column])
    print() 
```

```python
# Define the colors
colors = ['#0B4A71', '#048DC6', '#11B9EA', '#FDF5D2', '#A4A9AD']

# Function to wrap text
def wrap_labels(labels, width):
    return ['\n'.join(textwrap.wrap(label, width)) for label in labels]

# Create pie charts for each column
for column in df.columns:
    plt.figure(figsize=(3, 3))
    values = df[column].value_counts()
    labels = wrap_labels(values.index.tolist(), 30)
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.title(f'{column}')
    plt.ylabel('')  # Hide the y-label
    plt.show()
```
