```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Read data file
df = pd.read_excel('~/Downloads/data.xlsx')

# Print all column labels
print("Column labels:")
for column in df.columns:
    print(column)
```

```python
# Calculate correlation coefficients for the specified pairs
correlations = {
    'Age-Frequency': df['Age'].corr(df['Frequency']),
    'Age-Intensity': df['Age'].corr(df['Intensity']),
    'Gender-Frequency': df['Gender'].corr(df['Frequency']),
    'Gender-Intensity': df['Gender'].corr(df['Intensity']),
    'Gender-Safety': df['Gender'].corr(df['Safety']),
    'Frequency-Intensity': df['Frequency'].corr(df['Intensity']),
    'Frequency-Budget': df['Frequency'].corr(df['Budget']),
    'Intensity-Budget': df['Intensity'].corr(df['Budget'])
}

for pair, corr in correlations.items():
    print(f'Correlation between {pair}: {corr}')

# Convert the data into a DataFrame
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Pair', 'Correlation'])
correlation_df[['Variable1', 'Variable2']] = correlation_df['Pair'].str.split('-', expand=True)
correlation_df = correlation_df.pivot(index='Variable1', columns='Variable2', values='Correlation')

# Create a heatmap
plt.figure(figsize=(12, 5))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm', center=0)
plt.xticks(rotation=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# List of column names in your DataFrame
columns = df.columns.tolist()

# Create scatter plots for each pair of variables
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        plt.figure(figsize=(3, 3))
        plt.scatter(df[columns[i]], df[columns[j]])
        plt.xlabel(columns[i])
        plt.ylabel(columns[j])
        plt.title(f'{columns[i]} vs {columns[j]}')
        plt.show()
```
